```python
df=pd.read_sql_query("""
WITH  purchases_agg AS
(select 
p.PONumber,
count(distinct p.Brand) as total_brand,
sum(p.Quantity) as total_item_Quantity,
sum(p.Dollars) as total_item_Dollars,
avg(julianday(p.ReceivingDate)-julianday(p.PODate)) as avg_reciving_dely
from purchases p
group by p.PONumber)
                  
select 
v.PONumber,
v.Quantity as invoice_Quantity,
v.Dollars as invoice_Dollars,
v.Freight,
julianday(v.InvoiceDate)-julianday(v.PODate)   as day_po_to_invoice,             
julianday(v.PayDate)-julianday(v.InvoiceDate)  as day_to_pay,
pa.total_brand,
pa.total_item_Quantity,
pa.total_item_Dollars,
pa.avg_reciving_dely

from vendor_invoice as v
LEFT JOIN purchases_agg as pa
    on v.PONumber = pa.PONumber
                  """,cnn)
```

```python
def invice_risk(row):
    if (abs(row['invoice_Dollars']-row['total_item_Dollars'])>5):
        return 1
    
    if (row['avg_reciving_dely']>10):
        return 1
    
    return 0
df['flag_invoice']=df.apply(invice_risk,axis=1)
```

```python
sig_feat=[]
non_sig_feat=[]
result=[]
```

```python
metrix =['invoice_Quantity','invoice_Dollars',
         'Freight','day_po_to_invoice','day_to_pay',
         'total_brand','total_item_Quantity',
         'total_item_Dollars','avg_reciving_dely']
```

```python
from scipy.stats import ttest_ind
for Matrix in metrix:
    flaged_mean = flaged[Matrix].mean()
    noraml_mean = Normal[Matrix].mean()
    t_start,p_value = ttest_ind(
        flaged[Matrix].dropna(),
        Normal[Matrix].dropna(),
        equal_var=False
    )
    if p_value < 0.05:
        sig_feat.append(Matrix)
        result.append({
            'metrics':Matrix,
            'flaged_mean':flaged_mean.round(2),
            'normal_mean':noraml_mean.round(2),
            'P_value':p_value.round(3)
        })
    else:
        non_sig_feat.append(Matrix)
```

```python
def evel_model(true,predicted):
    accuracy_scor=accuracy_score(true,predicted)
    classification_repor=classification_report(true,predicted)
    precision_scor=precision_score(true,predicted)
    recall_scor=recall_score(true,predicted)
    f1_scor=f1_score(true,predicted)
    roc_auc_scor=roc_auc_score(true,predicted)
    return accuracy_scor,f1_scor,precision_scor,recall_scor,roc_auc_scor
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(probability=True),  # probability=True allows for predict_proba
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "CatBoost": CatBoostClassifier(verbose=0)  # verbose=0 keeps the output clean
}
```

```python
results_list=[]
```

```python
for model_name ,model in models.items():
    model.fit(X_train,Y_train)
    y_test_pred = model.predict(X_test)
    testacc_sc,testf1_scor,testprec_scor,testrecall_sc,testroc_scoe=evel_model(Y_test,y_test_pred)
    results_list.append([
        model_name, f"{testacc_sc:.4f}", f"{testf1_scor:.4f}", f"{testprec_scor:.4f}", f"{testrecall_sc:.4f}", f"{testroc_scoe:.4f}"
    ])


    print('Model preformance Test Prediction ')
    print('accuracy:{:.4f}'.format(testacc_sc))
    print('F1 score:{:.4f}'.format(testf1_scor))
    print('precision:{:.4f}'.format(testprec_scor))
    print('recall score:{:.4f}'.format(testrecall_sc))
    print('roc_auc_score:{:.4f}'.format(testroc_scoe))

    print('====================================================')
    print('\n')
 

```

```python
headers = ["Model", "Accuracy", "F1-Score", "Precision", "Recall", "ROC-AUC"]

# 3. Create the PDF using FPDF
class MLReportPDF(FPDF):
    def header(self):
        # Title Block
        self.set_font("Helvetica", "B", 16)
        self.set_text_color(26, 54, 93) # Deep Blue
        self.cell(0, 10, "Machine Learning Model Evaluation", ln=True, align="L")
        self.set_font("Helvetica", "", 10)
        self.set_text_color(100, 116, 139) # Muted Gray
        self.cell(0, 5, "Performance compilation across classification models.", ln=True, align="L")
        self.ln(10)

    def footer(self):
        self.set_y(-15)
        self.set_font("Helvetica", "I", 8)
        self.set_text_color(150, 150, 150)
        self.cell(0, 10, f"Page {self.page_no()}", align="C")

# Initialize PDF
pdf = MLReportPDF()
pdf.add_page()
pdf.set_font("Helvetica", "", 10)

# Set Column Widths (Must fit page printable area ~190mm wide)
col_widths = [45, 29, 29, 29, 29, 29]

# Render Table Header
pdf.set_font("Helvetica", "B", 10)
pdf.set_fill_color(43, 108, 176)  # Blue background
pdf.set_text_color(255, 255, 255) # White text
for i, header in enumerate(headers):
    pdf.cell(col_widths[i], 8, header, border=1, align="C", fill=True)
pdf.ln()

# Render Data Rows
pdf.set_font("Helvetica", "", 9)
pdf.set_text_color(0, 0, 0) # Black text

for row_idx, row in enumerate(results_list):
    # Zebra striping: Alternate light gray background
    if row_idx % 2 == 0:
        pdf.set_fill_color(247, 250, 252)
    else:
        pdf.set_fill_color(237, 242, 247)
        
    for col_idx, cell_data in enumerate(row):
        # Left align the model names, center align the numeric metrics
        align_mode = "L" if col_idx == 0 else "C"
        pdf.cell(col_widths[col_idx], 7, cell_data, border=1, align=align_mode, fill=True)
    pdf.ln()

# Save PDF to disk
target_dir = r"D:\Programing\Programing_folder\Program Files\Real World projects of data science\Invoice Intelligence ML\model_performance_report"

# Ensure the folder structure actually exists on your hard drive
os.makedirs(target_dir, exist_ok=True)

# Combine path with a specific filename
full_output_path = os.path.join(target_dir, "Model_test_Performance_Report.pdf")

# Export PDF file to the destination folder
pdf.output(full_output_path)
print(f"Successfully saved your metrics report to:\n{full_output_path}")
```

```python
RandomForest_parems = {
    'n_estimators': [100, 200, 300, 400, 500, 600, 700, 1500, 2000],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [1, 2, 3, 4, 5, 6, 7],
    'min_samples_split': [0.02, 0.1, 0.21, 0.3, 1.0], 
    'min_samples_leaf': [1, 2, 3, 4, 5, 6],
    'max_features': [1, 2, 3, 4, 5, 6]   
}
gradientboost_parem = {
    'loss': ['log_loss', 'exponential'], 
    'learning_rate': [0.1, 0.2, 0.25, 0.3, 1.0],
    'n_estimators': [100, 200, 300, 500],
    'subsample': [0.1, 0.2, 0.3, 0.4],
    'criterion': ['friedman_mse', 'squared_error'] 
}

Xg_parem = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [5, 6, 8, 10],
    'n_estimators': [100, 300, 500],
    'colsample_bytree': [0.1, 0.5, 0.8, 1.0]
}

KNN_parem = {
    'n_neighbors': [3, 5, 7, 9, 11, 15],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2] 
}

Catboost_parem = {
    'iterations': [100, 300, 500, 1000],
    'learning_rate': [0.01, 0.1, 0.2],
    'depth': [4, 6, 8, 10],
    'l2_leaf_reg': [1, 3, 5, 7, 9]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from catboost import CatBoostClassifier
from sklearn.base import BaseEstimator

# Patch CatBoost to be compatible with Scikit-Learn 1.6+
if not hasattr(CatBoostClassifier, "__sklearn_tags__"):
    CatBoostClassifier.__sklearn_tags__ = BaseEstimator.__sklearn_tags__
hyper_perm_models = [
    ('Random Forest', RandomForestClassifier(), RandomForest_parems),
    ('Gradient Boosting', GradientBoostingClassifier(), gradientboost_parem),
    ('XGBoost', XGBClassifier(), Xg_parem),
    ('KNN', KNeighborsClassifier(), KNN_parem),
    ('CatBoost', CatBoostClassifier(verbose=0), Catboost_parem)
]
```

```python
model_perms ={}


for model_name,model,model_perm in hyper_perm_models:
    current_n_jobs = None if model_name == 'CatBoost' else -1

    
    hyper_model = RandomizedSearchCV(estimator=model,
                                      param_distributions =model_perm,
                                      n_iter=5,
                                      verbose=1,
                                      n_jobs=current_n_jobs,
                                      cv=3
                                      )
    hyper_model.fit(X_train_scaled,Y_train)
    model_perms[model_name]=hyper_model.best_params_

for model_name in model_perms:
    print(f'---------------best parems {model_name}------------------')
    print(model_perms[model_name],'\n')
```

```python
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from catboost import CatBoostClassifier

# 1. Define the models with the exact best parameters discovered
final_models_config = [
    (
        'XGBoost', 
        XGBClassifier(
            n_estimators=500, 
            max_depth=10, 
            learning_rate=0.1, 
            colsample_bytree=1.0,
            n_jobs=-1,
            random_state=42
        )
    ),
    (
        'KNN', 
        KNeighborsClassifier(
            n_neighbors=5, 
            weights='uniform', 
            p=1, 
            algorithm='ball_tree',
            n_jobs=-1
        )
    ),
    (
        'CatBoost', 
        CatBoostClassifier(
            iterations=100, 
            depth=10, 
            learning_rate=0.2, 
            l2_leaf_reg=5,
            verbose=0,       # Keeps your notebook clean during training
            thread_count=-1  # Uses all CPU cores inside CatBoost safely
        )
    )
]

# 2. Train the models using the automated loop method
trained_models = {}

for model_name, model_instance in final_models_config:
    print(f"Training final {model_name} model on full training set...")
    
    # Train the model
    model_instance.fit(X_train_scaled, Y_train)
    
    # Save the fully trained model to your dictionary
    trained_models[model_name] = model_instance

print("\n All models successfully trained and saved!")
```

```python
results_list = []

for model_name, model in trained_models.items():
    print(f"\n{'='*40}")
    print(f"Evaluating {model_name}...")
    print(f"{'='*40}")
    
    # 1. Generate predictions on your test data
    y_pred = model.predict(x_test_scaled)
    
    # 2. Pass true labels and predictions to your evaluation function
    accuracy, f1, precision, recall, roc_auc = evel_model(Y_test, y_pred)
    
    # 3. Store the returned metrics for a final comparison table
    results_list.append({
        'Model': model_name,
        'Accuracy': round(accuracy, 4),
        'F1 Score': round(f1, 4),
        'Precision': round(precision, 4),
        'Recall': round(recall, 4),
        'ROC AUC': round(roc_auc, 4)
    })

# 4. Display a clean summary table of all models
print("\n--- Final Model Comparison ---")
results_df = pd.DataFrame(results_list)
print(results_df.to_string(index=False))
```
